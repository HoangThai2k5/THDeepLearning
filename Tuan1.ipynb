```python

f = np.array([1., 2., 3.])
g = np.array([4., 5., 6.])

# Xếp chồng theo chiều ngang (Horizontal Stack)
h_stack = np.hstack((f, g))
print('Horizontal Append\n', h_stack)


# Xếp chồng theo chiều dọc (Vertical Stack)
v_stack = np.vstack((f, g))
print('Vertical Append\n', v_stack)
```

```python


a = np.array([(1, 2, 3), (4, 5, 6)])
print(a)

# --- Slice theo hàng ---
print('first row:', a[0])   # Lấy hàng đầu tiên (chỉ số 0)
print('second row:', a[1])  # Lấy hàng thứ hai (chỉ số 1)

# --- Indexing & Slicing kết hợp ---
# Cấu trúc: a[hàng, cột]
# Dấu : đại diện cho "tất cả"

print('Second column:', a[:, 1])  # Lấy tất cả hàng, nhưng chỉ lấy cột thứ 2 (chỉ số 1)
print('first column:', a[:, 0])   # Lấy tất cả hàng, nhưng chỉ lấy cột đầu tiên (chỉ số 0)

# Lấy 2 giá trị đầu tiên của hàng thứ 2
print('First two values of second row:', a[1, :2])
```

```python
# Quy tắc: [số dòng, số cột]
a = np.array([(1, 2, 3), (4, 5, 6)])
print('first row:', a[0])
print('second row:', a[1])
print('Second column:', a[:, 1]) # Lấy tất cả dòng của cột chỉ số 1
print('first column:', a[:, 0])  # Lấy tất cả dòng của cột chỉ số 0
print('First two values of second row:', a[1, :2])
```

```python
import pandas as pd
# Tạo series từ list
a = pd.Series([1., 2., 3.])
print(a)

# Tạo series với index tùy chỉnh
b = pd.Series([1., 2., 3.], index=['a', 'b', 'c'])
print(b['a'])

# Tạo series từ dictionary
calories = {"day1": 420, "day2": 380, "day3": 390}
myvar = pd.Series(calories)
print(myvar)
```

```python
# Chuyển từ Numpy sang Pandas
h = np.array([[1, 2], [3, 4]])
df_h = pd.DataFrame(h)

# Chuyển từ Pandas ngược lại Numpy
numpy_h = np.array(df_h)

# Tạo DataFrame từ dictionary có tên cột
dic = {'Name': ['John', 'Smith'], 'Age': [30, 40]}
df_dic = pd.DataFrame(dic)

# Tạo DataFrame từ mảng ngẫu nhiên 20x4
r = np.random.randn(20, 4)
df = pd.DataFrame(r, columns=['so1', 'so2', 'so3', 'so4'])
print(df.head()) # In 5 dòng đầu tiên
```

```python


# Khởi tạo DataFrame thứ nhất
df1 = pd.DataFrame({
    'Name': ['John', 'Smith', 'Paul'],
    'Age': [25, 30, 50]
}, index=[0, 1, 2])

# Khởi tạo DataFrame thứ hai
df2 = pd.DataFrame({
    'Name': ['Adam', 'Smith'],
    'Age': [26, 11]
}, index=[3, 4])

# Nối hai DataFrame lại với nhau
df_concat = pd.concat([df1, df2])
print('Concate two DataFrame\n', df_concat)

# Xóa những giá trị ở cột Name trùng nhau (Giữ lại giá trị đầu tiên xuất hiện)
print('Drop duplicate values in DataFrame\n', df_concat.drop_duplicates('Name'))

# Sắp xếp DataFrame theo giá trị của cột 'Age' (mặc định là tăng dần)
print('Sort values in DataFrame\n', df_concat.sort_values('Age'))

# Đổi tên cột trong DataFrame
df_rename = df_concat.rename(columns={'Name': 'Surname', 'Age': 'Age_pupil'})
print('Rename columns of DataFrame\n', df_rename)
```

```python


# Đường dẫn đến tệp dữ liệu Fish.csv
path = '/content/sample_data/Fish.csv'

# Đọc dữ liệu từ CSV
df_csv = pd.read_csv(path, skipinitialspace=True, index_col=False)
print(df_csv.head())

# 1. Gom nhóm theo 'Species' và tính giá trị trung bình (mean) cho tất cả các cột số
print(df_csv.groupby(['Species']).mean().head())

# 2. Gom nhóm theo 'Species', nhưng chỉ lấy cột 'Weight' và tìm giá trị nhỏ nhất (min)
print(df_csv.groupby(['Species'])['Weight'].min().head())

# 3. Gom nhóm kết hợp theo nhiều cột ('Species' và 'Height'), sau đó tìm Weight nhỏ nhất
print(df_csv.groupby(['Species', 'Height'])['Weight'].min().head())
```

# Phần bài tập


```python
import numpy as np

# Bài 1 & 2: Khởi tạo mảng 3x3 ngẫu nhiên [0, 20]
arr = np.random.randint(0, 21, size=(3, 3))
print("Ma trận ngẫu nhiên:\n", arr)

# Bài 1: Kiểm tra tất cả phần tử có khác 0 không
check_all = np.all(arr != 0)
print("Tất cả phần tử khác 0:", check_all)

# Bài 2: Kiểm tra có tồn tại phần tử nào khác 0 không
check_any = np.any(arr != 0)
print("Có ít nhất một phần tử khác 0:", check_any)
```

```python
# Bài 3:
# Tạo mảng mẫu
a = np.array([1, 5, 10])
b = np.array([1, 2, 12])

print("Mảng a:", a)
print("Mảng b:", b)

print("a > b  (greater):      ", np.greater(a, b))
print("a >= b (greater_equal):", np.greater_equal(a, b))
print("a < b  (less):         ", np.less(a, b))
print("a <= b (less_equal):   ", np.less_equal(a, b))

# Tương tự cho mảng 2 chiều
matrix_a = np.array([[1, 2], [3, 4]])
matrix_b = np.array([[0, 2], [5, 4]])
print("\nSo sánh ma trận (greater):\n", np.greater(matrix_a, matrix_b))
```

```python
# Bài 4
# Tạo 10 số 1, 10 số 0 và 10 số 5
arr_1 = np.ones(10)
arr_0 = np.zeros(10)
arr_5 = np.ones(10) * 5

# Nối chúng lại thành một mảng duy nhất bằng hstack [cite: 171]
result_b4 = np.hstack((arr_1, arr_0, arr_5))
print("Mảng bài 4:\n", result_b4)
```

```python
# Bài 5
# Bắt đầu từ 30, kết thúc trước 71 để lấy được số 70
# Bước nhảy là 2 để chỉ lấy số chẵn
arr_even = np.arange(30, 71, 2)
print("Mảng số chẵn [30, 70]:\n", arr_even)
```

```python
# Bài 6
# Viết chương trình tạo ma trận đơn vị 3 x 3 (đường chéo chính bằng 1, các phần tử còn lại 0)
# Sử dụng hàm np.identity hoặc np.eye
identity_matrix = np.identity(3)
print("Ma trận đơn vị 3x3:\n", identity_matrix)
```

```python


# Bài 7: Mảng 10 phần tử [15, 55], in trừ phần tử đầu và cuối
arr7 = np.arange(15, 56) # Tạo dãy số [cite: 301]
# Lấy 10 phần tử đầu tiên để minh họa hoặc dùng linspace
arr7_10 = np.linspace(15, 55, 10)
print("Bài 7 - Các giá trị ở giữa:\n", arr7_10[1:-1]) # Sử dụng Slicing [cite: 319, 441]

```

```python
# Bài 8: Mảng 20 phần tử [0, 20], đổi dấu số trong khoảng [9, 15]
arr8 = np.arange(21)
arr8[(arr8 >= 9) & (arr8 <= 15)] *= -1 # Đổi dấu giá trị [cite: 442]
print("Bài 8 - Mảng sau khi đổi dấu:\n", arr8)
```

```python
# Bài 9: Ma trận 3x4 có giá trị trong khoảng [10, 21]
# Dùng randint để sinh số nguyên ngẫu nhiên [cite: 187]
arr9 = np.random.randint(10, 22, size=(3, 4))
print("Bài 9 - Ma trận 3x4:\n", arr9)
```

```python
# Bài 10: Ma trận 10x10, biên bằng 1, bên trong bằng 0
arr10 = np.ones((10, 10)) # Tạo ma trận toàn số 1 [cite: 68]
arr10[1:-1, 1:-1] = 0     # Gán vùng bên trong bằng 0
print("Bài 10 - Ma trận biên 1:\n", arr10)
```
