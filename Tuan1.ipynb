```python
import pandas as pd
# Tạo series từ list
a = pd.Series([1., 2., 3.])
print(a)

# Tạo series với index tùy chỉnh
b = pd.Series([1., 2., 3.], index=['a', 'b', 'c'])
print(b['a'])

# Tạo series từ dictionary
calories = {"day1": 420, "day2": 380, "day3": 390}
myvar = pd.Series(calories)
print(myvar)
```

```python
# Chuyển từ Numpy sang Pandas
h = np.array([[1, 2], [3, 4]])
df_h = pd.DataFrame(h)

# Chuyển từ Pandas ngược lại Numpy
numpy_h = np.array(df_h)

# Tạo DataFrame từ dictionary có tên cột
dic = {'Name': ['John', 'Smith'], 'Age': [30, 40]}
df_dic = pd.DataFrame(dic)

# Tạo DataFrame từ mảng ngẫu nhiên 20x4
r = np.random.randn(20, 4)
df = pd.DataFrame(r, columns=['so1', 'so2', 'so3', 'so4'])
print(df.head()) # In 5 dòng đầu tiên
```

```python


# Khởi tạo DataFrame thứ nhất
df1 = pd.DataFrame({
    'Name': ['John', 'Smith', 'Paul'],
    'Age': [25, 30, 50]
}, index=[0, 1, 2])

# Khởi tạo DataFrame thứ hai
df2 = pd.DataFrame({
    'Name': ['Adam', 'Smith'],
    'Age': [26, 11]
}, index=[3, 4])

# Nối hai DataFrame lại với nhau
df_concat = pd.concat([df1, df2])
print('Concate two DataFrame\n', df_concat)

# Xóa những giá trị ở cột Name trùng nhau (Giữ lại giá trị đầu tiên xuất hiện)
print('Drop duplicate values in DataFrame\n', df_concat.drop_duplicates('Name'))

# Sắp xếp DataFrame theo giá trị của cột 'Age' (mặc định là tăng dần)
print('Sort values in DataFrame\n', df_concat.sort_values('Age'))

# Đổi tên cột trong DataFrame
df_rename = df_concat.rename(columns={'Name': 'Surname', 'Age': 'Age_pupil'})
print('Rename columns of DataFrame\n', df_rename)
```

```python
import tensorflow as tf

@tf.function
def a_regular_function(x, y, b):
    x = tf.matmul(x, y)
    x = x + b
    return x

# Sử dụng hàm đã được chuyển sang Graph
x1 = tf.constant([[1.0, 2.0]])
y1 = tf.constant([[3.0], [4.0]])
b1 = tf.constant([10.0])
print(a_regular_function(x1, y1, b1))
```

```python
# Tạo một Layer đơn giản
class SimpleModule(tf.Module):
    def __init__(self, name=None):
        super().__init__(name=name)
        self.a_variable = tf.Variable(5.0, name="train_me")
        self.non_trainable_variable = tf.Variable(5.0, trainable=False, name="do_not_train_me")
    def __call__(self, x):
        return self.a_variable * x + self.non_trainable_variable

# Lưu checkpoint
my_model = SimpleModule(name="simple")
checkpoint = tf.train.Checkpoint(model=my_model)
checkpoint.write("my_checkpoint")

# Phục hồi checkpoint
new_model = SimpleModule()
new_checkpoint = tf.train.Checkpoint(model=new_model)
new_checkpoint.restore("my_checkpoint")
```

Tính toán trên Graph: Chuyển đổi các hàm Python thông thường sang môi trường Graph execution bằng @tf.function để tối ưu hóa tốc độ xử lý song song.  
Xây dựng Mô hình: Tìm hiểu cấu trúc của Modules, Layers và Dense để tạo ra các hàm toán học có khả năng học hỏi từ dữ liệu.  
Quản lý trạng thái: Sử dụng tf.Variable cho các giá trị có thể cập nhật và lưu trữ/phục hồi tiến trình huấn luyện thông qua Checkpoint.
